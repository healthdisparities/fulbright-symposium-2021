## Step 3. We will now look at the ancestry of individuals with AA allele.

### 1. Let's extract individuals from the VCF which have 0/0 (AA) allele for the varinat: rs1801133

````r
library("plyr")
library("dplyr")
library("ggplot2")
library("reshape2")
````

````r
#Let's read in the ancestry estimate file.
combinedAncestryEstimatesData = read.table("data/combinedAncestryEstimatesData.tsv", header = TRUE)
head(combinedAncestryEstimatesData)
````

````r
homozygousVarinatIndividuals = c("HG01112","HG01113","HG01119","HG01121","HG01122","HG01131","HG01134","HG01139","HG01140","HG01149","HG01256","HG01257",
                                 'HG01259',"HG01271","HG01275","HG01284","HG01342","HG01345","HG01350","HG01351","HG01353","HG01354","HG01372","HG01383",
                                 "HG01437","HG01440","HG01455","HG01474","HG01497")
homozygourIndividuals = combinedAncestryEstimatesData[combinedAncestryEstimatesData$IndividualID %in% homozygousVarinatIndividuals,]
head(homozygourIndividuals)
````

````r
homozygourIndividualsSorted = arrange(homozygourIndividuals, European, African, NativeAmerican, group_by = PopGroup)

row.names(homozygourIndividualsSorted) <- NULL
homozygourIndividualsSorted$index= as.numeric(rownames(homozygourIndividualsSorted))

homozygourIndividualsSortedMelt = melt(data = homozygourIndividualsSorted, id.vars = c("IndividualID", "PopGroup", "European", "African", "NativeAmerican", "index"), measure.vars = c("European", "African", "NativeAmerican"))

colnames(homozygourIndividualsSortedMelt)[7] <- 'Ancestry'
colnames(homozygourIndividualsSortedMelt)[8] <- 'AncestryFraction'
````

````r
colors  <- c("African" = "blue", "European" = "orange", "NativeAmerican" = "red")

options(repr.plot.width=16, repr.plot.height=8)
ggplot(data=homozygourIndividualsSortedMelt, aes(x=as.character(index), y=AncestryFraction, fill=Ancestry)) +
    geom_bar(stat="identity", width=1) + facet_grid(cols = vars(PopGroup), scales = "free", space = "free", drop = TRUE, switch="both") +
    scale_fill_manual(values=colors) + 
    labs(y="Percentage Ancestry Estimate", title = "Admixture plot for four population groups from 1KGP", subtitle = "Groups: IBS, PEL, YRI, & CLM") +
    theme(axis.title.x=element_blank(), axis.text.x=element_blank(), axis.ticks.x=element_blank(), panel.spacing.x=unit(1, "lines"), 
          strip.background = element_blank(), panel.background = element_blank(), axis.text=element_text(size=12),
          axis.title=element_text(size=18), strip.text.x = element_text(size = 18), title = element_text(size = 20))
````
